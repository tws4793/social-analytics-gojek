### Go-Jek

In this notebook, we are going to analyse the post-merger sentiments of **Go-Jek**.

We will look at the frequency of comments by week, as well as perform topic modelling based on sentiments from various review sites on Go-Jek.

## Frequency of Comments by Week

```python
df['week'] = df['date'].dt.week
df.head(10)
```

```python
week_count = df.groupby(['week']).count()
week_count['comment']
```

```python
week_count_re = df[df['source'] == 'reddit'].groupby(['week']).count()
week_count_re = week_count_re.reindex(week_count.index, fill_value=0)
week_count_re['comment']
```

```python
week_count_tw = df[df['source'] == 'twitter'].groupby(['week']).count()
week_count_tw = week_count_tw.reindex(week_count.index, fill_value=0)
week_count_tw['comment']
```

```python
x = week_count.index

# Initialisation
fig, ax = plt.subplots()
fig.set_dpi(300)

# Scatter Plot
ax.plot(x,week_count['comment'],color='g',label='Total')
ax.plot(x,week_count_re['comment'],color='r',label='Reddit')
ax.plot(x,week_count_tw['comment'],color='b',label='Twitter')

# Settings
ax.set_xticks(np.arange(1,x[-1]+2,2))
ax.set_xlim(1,x[-1])

ax.xaxis.set_tick_params(rotation=90)

ax.set_title('Frequency of Comments Concerning Go-Jek')
ax.set_xlabel('Week')
ax.set_ylabel('Comments')

ax.grid()
ax.legend()

plt.show()
```

## Topic Modeling

```python
def clean_list_tokenise(reviews_list):
    # Tokenise the words
    tokenizer = RegexpTokenizer(r'\w+')
    return [tokenizer.tokenize(sentence) for sentence in reviews_list]

def clean_list_lemma(reviews_list):
    # Lemmatization
    lemma = WordNetLemmatizer()
    return [[lemma.lemmatize(word) for word in sentence] for sentence in reviews_list]

def clean_list_stopwords(reviews_list, stop_other=[]):
    def stopword_condition(word):
        word = word.lower()
        
        return word not in stopwords.words('english') \
            and word not in stop_other \
            and word[:5] != 'http' \
            and word[:5] != 'https' \
            and word[:2] != 'RT' \
            and word[0] != '@'
    
    return [[w for w in s if stopword_condition(w)] for s in reviews_list]
```

```python
sentiment_filtered = clean_list_stopwords(clean_list_tokenise(df['comment'].tolist()))
len(sentiment_filtered)
```

```python
sentiment_filtered = clean_list_lemma(sentiment_filtered)
len(sentiment_filtered)
```

```python
dictlist = corpora.Dictionary(sentiment_filtered)
doc_term_matrix = [dictlist.doc2bow(doc) for doc in sentiment_filtered]

Lda = gensim.models.ldamodel.LdaModel
ldamodel = Lda(doc_term_matrix, num_topics=5, id2word=dictlist, passes=50)
pyLDAvis.gensim.prepare(ldamodel, doc_term_matrix, dictlist)
```
