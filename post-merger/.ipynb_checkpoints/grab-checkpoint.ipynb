## Sentiment Analysis

We will perform sentiment analysis on the comments. In this case, we will use vader to help us perform sentiment analysis on the comments. More readings on how sentiment analysis can be performed are available at:

- https://opensourceforu.com/2016/12/analysing-sentiments-nltk/
- http://t-redactyl.io/blog/2017/04/using-vader-to-handle-sentiment-analysis-with-social-media-text.html

```python
def analyse_sentiment_vader(df, col_name):
    sid = SentimentIntensityAnalyzer()
    vader = lambda text: sid.polarity_scores(text)
    
    df['vader'] = df[col_name].apply(vader)
    df = pd.merge(df, df['vader'].apply(pd.Series), left_index=True, right_index=True)
    return df.drop(['vader'], axis=1)
```

```python
df = analyse_sentiment_vader(df, 'comment')
df.iloc[np.r_[0:10, -10:0]]
```

### Average Sentiment - Across All Platforms

```python
df_period = df[df['date'] >= '2018-01-01']
```

```python
def common_graph_plot(ylabel, ylim_low,ylim_hi,title,grid=False):
    # Initialisation
    fig, ax = plt.subplots()
    fig.set_dpi(300)
    
    # Settings
    ax.set_ylim(ylim_low,ylim_hi)

    ax.xaxis.set_major_locator(mdates.MonthLocator())
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %y'))
    ax.xaxis.set_tick_params(rotation=270)

    ax.set_title(title)
    ax.set_xlabel('Time')
    ax.set_ylabel(ylabel)

    if grid:
        ax.grid()
    
    return fig, ax
```

```python
def plot_curve_fitting(df):
    x = df.index
    y = df['compound']
    
    x_dates = mdates.date2num(x)
    p3 = np.poly1d(np.polyfit(x_dates, y, 3))
    xx = np.linspace(x_dates.min(), x_dates.max(), len(x))
    return x_dates, p3(xx)
```

```python
fig, ax = common_graph_plot('Compound Score',-0.5, 0.5, \
              'Sentiment of Reviews of Grab',True)

avg_sentiment = df_period.groupby(['date']).mean()

x_dates, curve = plot_curve_fitting(avg_sentiment)
ax.plot(x_dates, curve, 'black', label='overall')

ax.legend()
plt.show()
```

```python
fig, ax = common_graph_plot('Compound Score',-0.5, 0.5, \
              'Sentiment of Reviews of Grab',True)

# colors = ['green','lightblue','blue','red']
colors = {
    'fb':'blue',
    'twitter':'lightblue',
    'gplay':'green',
    'reddit':'red'
}

for i, platform in enumerate(df_period['source'].unique()):
    dfp = df_period[df_period['source'] == platform]
    avg_sentiment_p = dfp.groupby(['date']).mean()
    
    x_dates, curve = plot_curve_fitting(avg_sentiment_p)
    ax.plot(x_dates, curve, colors[platform.lower()], label=platform.lower())

ax.legend()
plt.show()
```

## Examining the Drop in Sentiment with Topic Modelling

In most graphs, we observed that there has been a drop in sentiment between December 2017 and July 2018. What is causing this drop?

We turn towards topic modelling to see if it can help us explain why...

```python
df_study_period = df[df['date'].isin(pd.date_range('2017-12-01','2018-07-01'))]
df_study_period.iloc[np.r_[0:10, -10:0]]
```

### Cleaning

```python
def clean_list_tokenise(reviews_list):
    # Tokenise the words
    tokenizer = RegexpTokenizer(r'\w+')
    return [tokenizer.tokenize(sentence) for sentence in reviews_list]

def clean_list_lemma(reviews_list):
    # Lemmatization
    lemma = WordNetLemmatizer()
    return [[lemma.lemmatize(word) for word in sentence] for sentence in reviews_list]

def clean_list_stopwords(reviews_list, stop_other=[]):
    def stopword_condition(word):
        word = word.lower()
        
        return word not in stopwords.words('english') \
            and word not in stop_other \
            and word[:5] != 'http' \
            and word[:5] != 'https' \
            and word[:2] != 'RT' \
            and word[0] != '@'
    
    return [[w for w in s if stopword_condition(w)] for s in reviews_list]
```

```python
sentiment_filtered = clean_list_stopwords(clean_list_tokenise(list(df_study_period['comment'])))
len(sentiment_filtered)
```

```python
sentiment_filtered = clean_list_lemma(sentiment_filtered)
len(sentiment_filtered)
```

### Topic Modelling

```python
dictlist = corpora.Dictionary(sentiment_filtered)
doc_term_matrix = [dictlist.doc2bow(doc) for doc in sentiment_filtered]

Lda = gensim.models.ldamodel.LdaModel
ldamodel = Lda(doc_term_matrix, num_topics=5, id2word=dictlist, passes=50)
pyLDAvis.gensim.prepare(ldamodel, doc_term_matrix, dictlist)
```

## Analysis of the Results

The null hypothesis that Go-Jek wanted our team to prove was that the merger was causing a decline in sentiment.

Initially plotting a curve-fitting line across the period of July 2017 till November 2018, we observed that average sentiment for most platforms first fell sometime from December 2017 until mid-July 2018 before rising. However, the merger occurred on 26 March 2018, which means that the analyses does not fully answer the question of whether the merger was causing overall sentiment to fall. This is because even before the merger, it was already on the decline. Was it true that sentiment has been generally falling even before the merger? We decided to take a closer look at the period of December 2017 till July 2018 (the "observation period").

Investigating further using topic modelling for the observation period (set at 5 topics), we identified the possible topics (numbers correlate to the graph above):

1. Grab drivers
2. Grab app
3. General Ride-Hailing Industry
4. Promotions
5. Service related to Grab

Of interest is topic 2, which accounts for 24.9% of all words under study in the observation period. It seems that many people on social media were complaining about reliability issues with the Grab app itself. Users also seemed to complain about some Grab drivers in topic 1.

Taking into account all the results generated above, it is quite likely that the drop in sentiment between December 2017 and July 2018 were the result of Grab drivers and also the Grab app itself. Coincidentally, during the observation period, the Grab app crashed 3 times within a span of 2 months (on 1 March, 3 April and 17 April 2018). We cannot conclude that the merger was the sole seminal event causing the general decline in sentiment about Grab, but was caused by a confluence of other factors.

What could Go-Jek possibly take away from this? As much as Go-Jek hopes that they will not want to have a repeat of the PR disaster of Grab, controlling their drivers is something we believe is beyond their control. The next best thing Go-Jek could at least do is to ensure that their app (and the system overall) is reliable. Customers of ride hailing services these days demand a lot of convenience, particularly the fact that they should be able to quickly book rides within the shortest period of time.

## Limitations of the Analysis

It should be noted that these graphs are not topic specific and takes into account all factors relating to Grab, including but not related to app crashes or drivers cancelling on users.

One should also note that the original data does not accurately factor in location and context of the comments (apart from Reddit), and therefore as much as we wanted the source to originate from the Singaporean context, it is something that is beyond our control.

The analyses were based on simple average compound scores of each day, which means that such scores might be affected by outliers in the original sets of scores.
