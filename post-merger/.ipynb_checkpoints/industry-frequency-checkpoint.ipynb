```python
#Start the merger: 

path_csv = 'merged_all_platforms_v3.csv'
df = pd.read_csv(path_csv, encoding = 'utf-8', header = 0) 

#df = df.drop(columns = ['Unnamed: 0', "Unnamed: 0.1"])

#exclude all gplay data since we don't have the exact hours
df = df[~df['source'].str.contains("gplay") == True]

print("Rows: "+str(len(df)))
```

```python
df = df[~df['source'].str.contains("gplay") == True]

from dateutil import parser

df_new = pd.DataFrame()

def try_parse(date):
    try:
        date = parser.parse(date, yearfirst = True)
    except:
        date = None
        pass
    return date

dates = df['date'].apply(lambda date: try_parse(date))
dates = dates[dates.notnull()]
```

```python
def hr_func(date):
    return int(date.hour)

hours = dates.apply(lambda date: hr_func(date))
```

```python
import numpy as np
values, counts = np.unique(hours, return_counts=True)
print(counts)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [20, 10]
# data
df_data=pd.DataFrame({'values': values, 'counts': counts})
# plot

font = {'fontsize': 18, 'family': 'sans-serif'}

plt.plot( 'values', 'counts', data=df_data, linestyle='-', marker='o')
plt.xticks(np.arange(24), df_data['values'])
plt.grid(True, which = 'both')
plt.axis([0, 23, 8000, 22000])
plt.title('Frequency of Comments Concerning the Ride Hailing Industry in SG', fontdict = font)
plt.tick_params(labelsize=15)

plt.xlabel('Hour', fontdict=font)
plt.ylabel('Comments', fontdict=font)

plt.show()
#accross 3 SM platforms: FB, Twitter, Reddit (since GPlay does not contain times)
```

```python
df = pd.read_csv(path_csv, encoding = 'utf-8', header = 0) 
```

```python
df_twitter = df[df['source'].str.contains("Twitter") == True]
df_fb = df[df['source'].str.contains("Facebook") == True]
df_reddit = df[df['source'].str.contains("Reddit") == True]
df_gplay = df[df['source'].str.contains("gplay") == True]
```

```python
hours_twitter = df_twitter['date'].apply(lambda date: try_parse(date))
hours_twitter = hours_twitter[hours_twitter.notnull()]

hours_twitter = hours_twitter.apply(lambda date: hr_func(date))

hours_fb = df_fb['date'].apply(lambda date: try_parse(date))
hours_fb= hours_fb[hours_fb.notnull()]

hours_fb = hours_fb.apply(lambda date: hr_func(date))

hours_reddit = df_reddit['date'].apply(lambda date: try_parse(date))
hours_reddit = hours_reddit[hours_reddit.notnull()]

hours_reddit = hours_reddit.apply(lambda date: hr_func(date))
```

```python
values_twitter, counts_twitter = np.unique(hours_twitter, return_counts=True)
print(counts_twitter)
values_fb, counts_fb = np.unique(hours_fb, return_counts=True)
print(counts_fb)
values_reddit, counts_reddit = np.unique(hours_reddit, return_counts=True)
print(counts_reddit)
```

```python
df_all = pd.DataFrame()

df_all['hour'] = values_twitter
df_all['twitter'] = counts_twitter
df_all['fb'] = counts_fb
df_all['reddit'] = counts_reddit

df_all['twitter_%'] = df_all['twitter']/df_all['twitter'].sum()*100
df_all['fb_%'] = df_all['fb']/df_all['fb'].sum()*100
df_all['reddit_%'] = df_all['reddit']/df_all['reddit'].sum()*100

df_all
```

```python
plt.rcParams['figure.figsize'] = [25, 10]

plt.plot( 'hour', 'twitter_%', data=df_all, linestyle='-', marker='o', c = '#0084b4')
plt.plot( 'hour', 'fb_%', data=df_all, linestyle='-', marker='o', c = '#3B5998')
plt.plot( 'hour', 'reddit_%', data=df_all, linestyle='-', marker='o', c='#ff4500', alpha = 0.5)
#excluding Reddit - not enough data

plt.xticks(np.arange(24), df_all['hour'])
plt.grid(True, which = 'both')

plt.axis([0, 23, 0, 10])
plt.title('% of Comments Posted Each Hour - Facebook, Twitter, Reddit', fontdict = font)
plt.tick_params(labelsize=15)

plt.legend(['Twitter', 'Facebook', 'Reddit'], prop={'size': 20})

plt.xlabel('Hour', fontdict=font)
plt.ylabel('% of Comments on Given Platform', fontdict=font)

plt.show()
```
